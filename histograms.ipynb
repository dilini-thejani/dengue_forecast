```python
import numpy as np  # Import the NumPy library for numerical operations
import pandas as pd  # Import the Pandas library for data manipulation
import matplotlib.pyplot as plt  # Import the Matplotlib library for creating plots
```

```python
# Read a CSV file named "pro-2010.csv" into a Pandas DataFrame and store it in the variable df10
df10 = pd.read_csv("/content/drive/My Drive/Datasets/Dengue/Pro-2010.csv")

# Extract the 'Total' column from the DataFrame and store it in the variable Total
Total = df10.Total

# Create a histogram using the data in the 'Total' variable with blue bars
plt.hist(Total, color='b')

# Add a label to the x-axis of the histogram
plt.xlabel("Dengue Incidence", fontsize=16)

# Add a label to the y-axis of the histogram
plt.ylabel("Frequency", fontsize=16)

# Add a title to the histogram
plt.title("2010", fontsize=20)

# Display the histogram plot
plt.show()
plt.savefig("2010")
```
