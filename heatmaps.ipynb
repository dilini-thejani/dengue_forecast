```python
import pandas as pd  # Import the Pandas library for data manipulation
import matplotlib.pyplot as plt  # Import the Matplotlib library for creating plots
import seaborn as sns  # Import the Seaborn library for data visualization
```

```python
file_path1 = '/content/drive/My Drive/Datasets/Dengue/Pro-2010.csv'
file_path2 = '/content/drive/My Drive/Datasets/Climate/2010-Precipitation.csv'
file_path3 = '/content/drive/My Drive/Datasets/Climate/2010-Temperature.csv'
file_path4 = '/content/drive/My Drive/Datasets/Climate/2010-Humidity.csv'

# Read data from CSV files into Pandas DataFrames
df10 = pd.read_csv(file_path1)
df10p = pd.read_csv(file_path2)
df10t = pd.read_csv(file_path3)
df10h = pd.read_csv(file_path4)

# Rename columns in each DataFrame for clarity
df2010 = df10.rename(columns={"Total": "Dengue"})
df2010p = df10p.rename(columns={"Annual": "Prec."})
df2010t = df10t.rename(columns={"Annual": "Tem."})
df2010h = df10h.rename(columns={"Annual": "Hum."})

# Concatenate the selected columns from different DataFrames into a single DataFrame
Df10 = pd.concat([df2010['Dengue'], df2010p['Prec.'], df2010t['Tem.'], df2010h['Hum.']], axis=1)

# Create a correlation heatmap using Spearman correlation method
sns.heatmap(Df10.corr(method="spearman"), annot=True, square=True, cmap="coolwarm")

# Add a title to the heatmap
plt.title('2010')
plt.show()
```
